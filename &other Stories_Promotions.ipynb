```python
BASE_PATH = "https://www.stories.com/en_sek/index.html"
```

```python
page = requests.get(BASE_PATH)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
menu = soup.findAll("ul",{"class": "categories"})
```

```python
menu
```

```python
for tab in menu:
    print(tab.text)
```

```python
for tab in menu:
    print(tab.text.strip())
```
